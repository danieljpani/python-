```
# Mini ETL Project
import requests
url='https://restcountries.com/v3.1/all'
data=requests.get(url).json()
countries=[]
for c in data:
    countries.append({
        'name': c.get('name',{}).get('common','Unknown'),
        'region': c.get('region','Unknown'),
        'population': c.get('population',0),
        'area_km2': c.get('area',0)
    })
etl_df=pd.DataFrame(countries)
etl_df.to_csv('countries_clean.csv',index=False)
print(etl_df.head())
```
